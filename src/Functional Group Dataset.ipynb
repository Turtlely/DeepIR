```python
import pandas as pd
import config
```

```python
#Smiles csv file
SMILES_path = config.ROOT_DIR+'/Datasets/SMILES.csv'

#Import SMILES file
data = pd.read_csv(SMILES_path,index_col=0,sep='\t')
```

```python
data
```

```python
from rdkit import Chem
import numpy as np
```

```python
from rdkit.Chem import Fragments

func_grp_smarts = {'alkane':'[CX4;H0,H1,H2,H4]','methyl':'[CH3]','alkene':'[CX3]=[CX3]','alkyne':'[CX2]#C',
                   'alcohols':'[#6][OX2H]'}

groups = []

for index,row in data.iterrows():
    SMILES = row[1]

    try:
        mol = Chem.MolFromSmiles(SMILES)

        #Maybe put the stuff below into a function
        
        #Substructure search for the functional groups
        
        '''Essential functional groups''' #missing alcohols
        #Many of these are redundant. For example, any imide will also have a tertiary amine
        
        COOH = Fragments.fr_COO(mol) #Carboxylic acid
        CO = Fragments.fr_C_O(mol) #Carbonyls
        CS = Fragments.fr_C_S(mol) #Thiocarbonyls
        Imines = Fragments.fr_Imine(mol) #Imines
        Imides = Fragments.fr_imide(mol) #number of imides
        NH2 = Fragments.fr_NH2(mol) #Primary amines
        NH1 = Fragments.fr_NH1(mol) #Secondary amines
        NH0 = Fragments.fr_NH0(mol) #Tertiary amines
        amides = Fragments.fr_amide(mol) #number of amides
        SH = Fragments.fr_SH(mol) #Thiol groups
        Aldehydes = Fragments.fr_aldehyde(mol) #Aldehydes
        azide = Fragments.fr_azide(mol) #number of azides
        azo = Fragments.fr_azo(mol) #number of azo groups
        phenol = Fragments.fr_phenol(mol) #number of phenols
        benzene = Fragments.fr_benzene(mol) #number of benzenes
        ester = Fragments.fr_ester(mol) #number of esters
        ether = Fragments.fr_ether(mol) #number of ethers
        urea = Fragments.fr_urea(mol) #number of urea groups
        ketones = Fragments.fr_ketone(mol) #number of ketones
        nitrile = Fragments.fr_nitrile(mol) #number of nitrile groups
        nitro = Fragments.fr_nitro(mol) #number of nitro groups
        nitrobenzene = Fragments.fr_nitro_arom(mol) #number of nitrobenzenes
        nitroso = Fragments.fr_nitroso(mol) #number of nitroso groups
        methyoxy = Fragments.fr_methoxy(mol) #number of methoxy groups
        CX = Fragments.fr_alkyl_halide(mol) #number of alkyl halides
        epoxides = Fragments.fr_epoxide(mol) #number of epoxides
        
        extragroups = []
        '''Functional groups not found by RDkit natively'''
        for f in func_grp_smarts.items():
            #print(f[1])
            extragroups.append(len(mol.GetSubstructMatches(Chem.MolFromSmarts(func_grp_smarts[f[0]]))))
        
        '''Generally more complex functional groups'''
        #Omitted from the dataset for preliminary training
        '''
        halogens = Fragments.fr_halogen(mol) #number of halogens
        Pyrroles = Fragments.fr_Nhpyrrole #number of pyrroles
        Alkyl_Carbamate = Fragments.fr_alkyl_carbamate(mol) #number of alkyl carbamates
        amidine = Fragments.fr_amidine(mol) #number of amidines
        aniline = Fragments.fr_aniline(mol) #number of anilines
        barbituates = Fragments.fr_barbitur(mol) #number of barbituate groups
        furan = Fragments.fr_furan(mol) #number of furans
        guanidine = Fragments.fr_guanido(mol) #number of guanidines
        hydrazine = Fragments.fr_hdrzine(mol) #number of hydrazines
        imidazole = Fragments.fr_imidazole(mol) #number of imidazole
        isocyan = Fragments.fr_isocyan(mol) #number of isocyanates
        isothiocyan = Fragments.fr_isothiocyan(mol) #number of isothiocyanates
        betalactam = Fragments.fr_lactam(mol) #number of beta lactams
        lactone = Fragments.fr_lactone(mol) #number of lactones
        morpholine = Fragments.fr_morpholine(mol) #number of morpholine groups
        oxazole = Fragments.fr_oxazole(mol) #number of oxazole rings
        oxime = Fragments.fr_oxime(mol) #number of oxime groups
        pAcid = Fragments.fr_phos_acid(mol) #number of phosphoric acid groups
        pEster = Fragments.fr_phos_ester(mol) #number of phosphoric ester groups
        piperzine = Fragments.fr_piperzine(mol) #number of piperzine rings        
        piperdine = Fragments.fr_piperdine(mol) #number of piperdine rings
        prisulfonamid = Fragments.fr_prisulfonamd(mol) #number of primary sulfonamides
        pyridine = Fragments.fr_pyridine(mol) #number of pyridines
        thioether = Fragments.fr_sulfide(mol) #Number of thioethers
        sulfonamides = Fragments.fr_sulfonamd(mol) #number of sulfonamide groups
        sulfone = Fragments.fr_sulfone(mol) #number of sulfone groups
        tetrazole = Fragments.fr_tetrazole(mol) #number of tetrazole rings
        thiazole = Fragments.fr_thiazole(mol) #Number of thiazole rings
        thiocyanates = Fragments.fr_thiocyan(mol) #number of thiocyanates
        thiophene = Fragments.fr_thiophene(mol) #number of thiophenes'''
        
        func_groups_common = [COOH,CO,CS,Imines,Imides,NH2,
                              NH1,NH0,amides,SH,Aldehydes,azide,
                              azo,phenol,benzene,ester,ether,urea,
                              ketones,nitrile,nitro,nitrobenzene,nitroso,methyoxy,
                              CX,epoxides]
        
        '''[COOH,CO,CS,Imines,Imides,NH2,
            NH1,NH0,amides,SH,Aldehydes,azide,
            azo,phenol,benzene,ester,ether,urea,
            ketones,nitrile,nitro,nitrobenzene,nitroso,methyoxy,
            CX,epoxides,alkane,methyl,alkene,alkyne,alcohols]'''

        datarow = [row[0]]+func_groups_common + extragroups
        groups.append(datarow)
        #print(SMILES,str(func_groups_common))
        
    except Exception as e:
        groups.append([np.NaN]*32)
        continue    
```

```python
print(len(groups[0]))
GroupData = pd.DataFrame(groups,columns=['CAS','COOH','CO','CS','imines','imides','NH2','NH1','NH0','amides','SH','aldehyde','azide','azo','phenol','benzene','ester','ether','urea','ketones','nitrile','nitro','nitrobenzene','nitroso','methoxy','CX','epoxides','alkane','methyl','alkene','alkyne','alcohols'])
```
