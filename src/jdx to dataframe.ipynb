# Convert JDX files into a dataframe

```python
import pandas as pd
from jcamp import JCAMP_calc_xsec, JCAMP_reader
import os
import numpy as np
from scipy.interpolate import interp1d
import config
```

```python
IR_dir = config.ROOT_DIR'/Datasets/IR_spectra/'
rows = []
```

```python
for index,file in enumerate(os.listdir(IR_dir)):
    try:
        print(index,file)
        data = JCAMP_reader(IR_dir+file)

        #Spectra processing
        JCAMP_calc_xsec(data)

        #CAS ID scraping
        CAS = data['cas registry no']
        
        #Absorbance spectrum
        spectrum =[]

        #Interpolate spectra between 600 and 3500cm^-1
        x = np.linspace(600,3500,num=12000)

        if(data['yunits'].lower()=='absorbance'):
            f = interp1d(data['wavenumbers'],data['y'])
            spectrum = f(x)

        elif (data['yunits'].lower() == 'transmittance'):
            f = interp1d(data['wavenumbers'],data['absorbance'])
            spectrum = f(x)
        else:
            print("Incompatible units")
            continue
        
        test = [CAS]+list(spectrum)
        rows.append(test)
    except Exception as e:
        #Sometimes interpolation will fail, we dont want these
        print(e)
```

```python
IRFrame = pd.DataFrame()
IRFrame = IRFrame.append(rows)
```

# Dataset creation
First column is the CAS number, the next 12000 columns are IR spectra data
