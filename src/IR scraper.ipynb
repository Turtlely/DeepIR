```python
import pandas as pd
import config
```

```python
species_path = config.ROOT_DIR+'/Datasets/species.txt'
species_list = pd.read_csv(species_path,sep='\t')
species_list.columns = ["NAME", "FORMULA", "CAS"]
species_list = species_list.dropna()
```

```python
print(species_list)
```

# Download IR data

```python
import requests

NIST_URL = "https://webbook.nist.gov/cgi/cbook.cgi"
CAS_LIST = species_list['CAS']
NAME_LIST = species_list['NAME']

IR_dir = config.ROOT_DIR+'/Datasets/IR_spectra/'

for index,cas_id in enumerate(CAS_LIST[13074:],start=13074):
    params={'JCAMP': 'C'+cas_id, 'Type': 'IR', 'Index': 0}	
    response = requests.get(NIST_URL, params=params)

    if response.text == '##TITLE=Spectrum not found.\n##END=\n':
        continue
    
    with open(IR_dir+cas_id,'wb') as spec:
        spec.write(response.content)
    print(index)
```
