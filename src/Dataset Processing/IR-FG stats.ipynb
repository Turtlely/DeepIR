Function group statistics

```python
import matplotlib.pyplot as plt
import numpy as np
group_counts = []

for g in list(dataset.columns[12000:]):
    group = dataset[g].value_counts()
    group_counts.append(np.array(group))
```

```python
labels = ['COOH','CO','CS','imines','imides','NH2','NH1','NH0','amides','SH','aldehyde','azide','azo','phenol','benzene','ester','ether','urea','ketones','nitrile','nitro','nitrobenzene','nitroso','methoxy','CX','epoxides','alkane','methyl','alkene','alkyne','alcohols']
df = pd.DataFrame(group_counts,index=labels)
df
```

```python
import config

for i in range(31):
    fig, ax = plt.subplots()
    index = i
    df.iloc[index].dropna().plot.bar(figsize=(10,5))
    ax.set_title(labels[index])
    ax.set_yscale('log')
    plt.xticks(rotation=0)
    plt.grid(True)
    plt.savefig(config.ROOT_DIR+'/Datasets/plots/'+labels[index]+'_stats.png')
```
