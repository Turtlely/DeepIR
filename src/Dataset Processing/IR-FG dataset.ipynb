```python
import pandas as pd
import numpy as np
import config
```

```python
IR = pd.read_csv(config.ROOT_DIR+'/Datasets/IR.csv', sep='\t',index_col=0)
FG = pd.read_csv(config.ROOT_DIR+'/Datasets/functional_groups.csv',index_col=0,sep='\t')
FG.dropna()
```

```python
IR
```

```python
new_rows = []
```

```python
for index,row in IR.iterrows():
    try:
        CAS_id=row[0]
        #print(CAS_id)
        IR_array = row.values[1:]
        x = FG.loc[FG['CAS']==CAS_id]
        if len(x.values) == 0:
            continue
        print(index)
        FG_array = x.values[0][1:]
        new_array = np.append(IR_array,FG_array)
        
        new_rows.append(new_array.tolist())
        
    except Exception as e:
        print(e)
```

```python
print(len(new_rows))
```

```python
IR_FG = pd.DataFrame(new_rows)
```

```python
IR_FG
```

```python
IR_FG.to_csv(config.ROOT_DIR+'/Datasets/IR-FG.csv',sep='\t')
```
