```python
import pandas as pd
import config
```

```python
species_path = config.ROOT_DIR+'/Datasets/species.txt'
species_list = pd.read_csv(species_path,sep='\t')
species_list.columns = ["NAME", "FORMULA", "CAS"]
species_list = species_list.dropna()
```

```python
print(species_list)
```

```python
import pubchempy as pcp
import numpy as np

NIST_URL = "https://webbook.nist.gov/cgi/cbook.cgi"
CAS_LIST = species_list['CAS']
NAME_LIST = species_list['NAME']

IR_dir = '/Users/ryantang/Documents/GitHub/DeepIR/Datasets/IR_spectra/'

entries = []

for index,cas_id in enumerate(CAS_LIST):
    try:
        c = pcp.get_properties('CanonicalSMILES',cas_id,'name')
        print(c,index)
        entries.append((cas_id,c[0]['CanonicalSMILES']))
    except IndexError:
        entries.append((cas_id,np.NaN))

```

```python
SMILES_frame = pd.DataFrame(entries,columns=['CAS','CSMILES'])
```
